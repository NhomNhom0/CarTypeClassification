```python
import os
import numpy as np
from PIL import Image
import pandas as pd
```

```python
SUV_folder = ["/kaggle/input/car-types-recognition/SUV/SUV",
              "/kaggle/input/stanford-car-body-type-data/stanford_cars_type/SUV"]
SUV_path = []

for folder in SUV_folder:
    file_names = os.listdir(folder)
    sorted_file_names = sorted(file_names, key=lambda x: int(''.join(filter(str.isdigit, x))))
    image_paths = [os.path.join(folder, filename) for filename in sorted_file_names]
    SUV_path.extend(image_paths)

SUV_label = [1] * len(SUV_path)
```

```python
cars_file = []
cars_file.extend(SUV_path)
len(cars_file)
```

```python
cars_label = []
cars_label.extend(SUV_label)
len(cars_label)
```

```python
def AddData(car_folder, car_type):
    car_path = []
    for folder in car_folder:
        file_names = os.listdir(folder)
        sorted_file_names = sorted(file_names, key=lambda x: int(''.join(filter(str.isdigit, x))))
        image_paths = [os.path.join(folder, filename) for filename in sorted_file_names]
        car_path.extend(image_paths)

    car_filenames = [os.path.basename(path) for path in car_path]

    car_label = [car_type] * len(car_filenames)
    
    cars_label.extend(car_label)
    cars_file.extend(car_path)
```

```python
Sedan_folder = ["/kaggle/input/car-types-recognition/sedan/sedan", 
               "/kaggle/input/stanford-car-body-type-data/stanford_cars_type/Sedan"]
AddData(Sedan_folder, 2)

print(len(cars_file))
print(len(cars_label))
```

```python
Convertible_folder = ["/kaggle/input/car-types-recognition/convertable/convertable",
                     "/kaggle/input/stanford-car-body-type-data/stanford_cars_type/Convertible"]
AddData(Convertible_folder, 3)

print(len(cars_file))
print(len(cars_label))
```

```python
Wagon_folder = ["/kaggle/input/stanford-car-body-type-data/stanford_cars_type/Wagon",
               "/kaggle/input/car-types-recognition/station_wagon/station_wagon"]

AddData(Wagon_folder, 4)

print(len(cars_file))
print(len(cars_label))
```
